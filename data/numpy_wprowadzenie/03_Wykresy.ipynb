```python
import numpy as np
import matplotlib.pyplot as plt
# wykresy będa się renderowały w notebooku
%matplotlib inline  
```

```python
# podajemy liczby x, y i parametry dodatkowe
plt.plot([1,2,3,4], [5,7,9,3], 'go')
plt.axis([0,10, -10,20])  #podajemy zakresy min-max osi x oraz y
plt.title("Tytuł tego wykresu")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
```

```python
# wykres z wieloma seriami
x= np.arange(0, 5.2, 0.2)
plt.figure(figsize=(10,6)) 
plt.plot(x, x, 'g--')
plt.plot(x, x**2, 'r.-')
plt.plot(x, 1/3*x**3, 'k-.')
plt.text(0, 35, 'Tekst na wykresie', fontsize=20, color='fuchsia')
plt.text(0, 30, 'Tekst na wykresie', fontsize=20, color='magenta')
arrow={"color":"blue", "linewidth":1, "arrowstyle": '->'}
plt.annotate('Adnotacja na wykresie', [2,5], [0.5, 20], arrowprops = arrow )
plt.show()
```

```python
plt.plot(x,x,'g--', x, x**2, 'ro')
```

```python
srednia, odchylenie = 100, 15
np.random.seed(0)
x= srednia + odchylenie * np.random.randn(10000)
plt.hist(x, bins = 30, color = 'green', alpha=0.5)
plt.grid(axis='y')
plt.xticks([60,80,100,120,140])
plt.show()
```

```python
kraje = ['USA', 'RUS', 'GER']
m_brazowy = np.array([12,10,8])
m_srebrny = np.array([4,3,2])
m_zloty = np.array([3,1,2])

plt.bar(kraje, m_brazowy, color='brown', width=0.5, label = 'B')
plt.bar(kraje, m_srebrny, color='silver', width=0.5, label = 'S', bottom=m_brazowy)
plt.bar(kraje, m_zloty, color='gold', width=0.5, label = 'Z', bottom=m_brazowy+m_srebrny)

plt.legend(ncol=3, loc='best')

plt.show()
```
