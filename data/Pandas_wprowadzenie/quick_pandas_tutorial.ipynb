## Filtrowanie

```python
# spojniki logiczne
# or:|  and:&  not:~
df[ (df['Lscore']>100)  & (df['Lscore']<200) ]
```

```python
# sezony 88-90 Wloc = N, LScore< 95 Wscore>155
df.describe()
```

```python
# warunki oddzielnie
sezony_88_90 = df.Season.isin([1988, 1989, 1990])
#wloc_n = df.Wloc=='N'
wloc_n = df.Wloc.str.startswith("N")
lscore_lt_95 = df.Lscore<95
wscore_gt_155 = df.Wscore>155

# filtrowanie
df[sezony_88_90 & wloc_n & lscore_lt_95 & wscore_gt_155]
```

## Grupowanie

```python
df = pd.read_csv("RegularSeasonCompactResults.csv")
df.Wloc.value_counts()
```

```python
df.groupby("Wloc")["Wscore"].mean()
```

```python
# pogrupuj po Season i Wloc i wypisz agregacjÄ™ Wscore = max, Lscore = min, nowa = mean
#df.groupby("Wloc", "nowa")["Wscore"].mean()

df.groupby(["Season", "Wloc"]).agg({
    "Wscore": np.max,
    "Lscore": np.min,
    "nowa": np.mean
})
```
